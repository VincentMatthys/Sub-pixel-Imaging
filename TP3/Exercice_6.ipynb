```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Global parameter
N = 100
```

```python
X = np.linspace(-N/2, N/2, 10001)
```

```python
# Sinus cardinal
Y1 = np.sin(np.pi * X) / (np.pi * X)
Y1[X == 0] = 1
# Sinus cardinal discret
if N % 2 == 0:
    Y2 = np.sin(np.pi * X) / (N * np.sin(np.pi * X / N))
#     Y2[X == X.astype(int)] = 1
else :
    Y2 = np.sin(np.pi * X) / (N * np.tan(np.pi * X / N))
#     Y2[X == X.astype(int)] = 1
plt.figure(figsize = (20, 10))
plt.xticks(fontsize = 20);
plt.yticks(fontsize = 20);
plt.plot(X, Y2, 'r', label = "sincd");
plt.plot(X, Y1, label = "sinc");
plt.ylim(-0.1, 0.1)
legend = plt.legend(fontsize = 30)
frame = legend.get_frame()
frame.set_facecolor('blue')
frame.set_linewidth(3)
frame.set_alpha(0.3)
plt.xlabel("x", fontsize= 20)
plt.ylabel("y", fontsize= 20)
plt.grid(color='gray', linestyle='--', linewidth=1)
```
